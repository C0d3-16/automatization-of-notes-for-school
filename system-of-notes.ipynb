```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
```

```python
doc = DocxTemplate("example-of-notes.docx")
```

```python
TeacherName = "Diego Garcia"
NumberTelephone = "123456789(0)"
Mail = "DiegoGarcia321@edu.co"
NumberOfClass = "8-2"
SchoolYear = "2024"
Date = datetime.today().strftime("%d/%m/%Y")
```

```python
print(Date)
```

```python
constantes = {'TeacherName': TeacherName, 'NumberTelephone': NumberTelephone, 'Mail': Mail,
              'Date': Date, 'SchoolYear': SchoolYear, 'NumberOfClass': NumberOfClass}
```

```python
print(constantes)
```

```python
df = pd.read_excel('Students.xlsx')
```

```python
df
```

```python
for indice, fila in df.iterrows():
    contenido = {
        'NameofStudent':fila["NameofStudent"],
        'Mat1': fila['Matematica'],
        'Fis1': fila['Fisica'],
        'Qui1': fila['Quimica'],
        'DaysOn': fila['DaysOn'],
        'DaysOff': fila['DaysOff'],
    }

  
    contenido.update(constantes)
    doc.render(contenido)
    doc.save(f"Notas_de_{fila['NameofStudent']}.docx")
    

    print(contenido)
```

```python
doc.render(constantes)
doc.save(f"prueba.docx")
```
